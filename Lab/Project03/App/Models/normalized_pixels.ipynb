# **I &nbsp;&nbsp;&nbsp; Import Libraries**

```python
import numpy as np
import pickle
from Utils.mnist import MNIST
from Utils.softmax_regression import SoftmaxRegression
from Utils.feature_vector import FeatureVector
```

# **II &nbsp;&nbsp;&nbsp; Load Data**

```python
(X_train, y_train), (X_test, y_test) = MNIST().load_data()
```

# **III &nbsp;&nbsp;&nbsp; Experiments**

## **1 &nbsp;&nbsp;&nbsp; Data Preprocessing**

```python
feature_vector = FeatureVector()
num_classes = 10
```

```python
X_train = feature_vector.normalized_pixels(X_train)
y_train = np.eye(num_classes)[y_train]

X_test = feature_vector.normalized_pixels(X_test)
y_test = np.eye(num_classes)[y_test]
```

## **2 &nbsp;&nbsp;&nbsp; Train Model**

```python
model = SoftmaxRegression(learning_rate=0.1, epochs=100, batch_size=64)
model.fit(X_train, y_train)
```

```python
model.plot_loss()
```

## **3 &nbsp;&nbsp;&nbsp; Evaluate Model**

```python
model.eval(X_test, y_test, average="none")
```
